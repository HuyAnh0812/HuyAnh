```python
import math
def sochinhphuong(n):
    căn=math.isqrt(n)
    if căn*căn==n :
        return True
    else :
        return False
n = int(input())
print(sochinhphuong(n))
```

```python
import math
def perfect(n):
    tong = 1
    for i in range (2,math.isqrt(n)+1) :
        if n%i ==0 :
            tong +=i
            if i != n//i :
                tong += n//i
    return tong == n
n = int(input())
print(perfect(n))
```

```python
import math 
n = int(input())
def prime(n):
    if n < 2 :
        return False
    for i in range (2,math.isqrt(n)+1):
        if n%i==0:
            return False
        return True
dem=0
while n>0 :
    n//=10
    if prime(n):
        dem+=1
print(dem)
```
