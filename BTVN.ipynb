```python
#6
a, b, c = map(int, input().split())
p = (a + b + c)/2
s = (p*(p-a)*(p-b)*(p-c))**1/2
if a + b > c :
    print('%.1f' %s)
elif a + c > b :
    print('%.1f' %s)
elif b + c > a :
    print('%.1f' %s)
else :
    print('khong phai ba canh cua tam giac')
```

```python
#8
a = input('Nhập tên chủ hộ')
b = int(input('Nhập chỉ số điện kế tháng trước'))
c = int(input('Nhập chỉ số điện kế tháng này'))
f = c - b
tiendien = 0
if 0 < f <= 50 :
    tiendien = f * 1984
elif 51 <= f <= 100 :
    tiendien = 50 * 1984 + 2050 * (f-50)
elif 101 <= f <= 200 :
    tiendien = 50 * 1984 + 2050 * 50 + 2380 * (f-100)
elif 201 <= f <= 300 :
    tiendien = 50 * 1984 + 50 * 2050 + 100 * 2380  + 2998 * (f-200)
elif 301 <= f <= 400 :
    tiendien = 50 * 1984 + 50 * 2050 + 2380 * 100 + 2998 * 100 + 3350 * (f-300)
elif f >= 400 :
    tiendien = 50 * 1984 + 50 * 2050 +  2380 * 100  + 2998 * 100 + 100 * 3350 + 3460 * (f-400)
tienvat = tiendien * 0.08
tiendientong = tienvat + tiendien
print('tiền vat', int(round(tienvat)))
print('tiền điện tổng phải trả là :', int(round(tiendientong)))

    
```
