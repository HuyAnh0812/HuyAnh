```python
#3
a, b, c = map(float, input("Nhập 3 cạnh a, b, c: ").split())

if a + b > c and a + c > b and b + c > a:
    print("Đây là tam giác.", end=' ')
    if a == b == c:
        print("Loại: Tam giác đều.")
    elif a == b or b == c or a == c:
        print("Loại: Tam giác cân." if not (a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2) else "Loại: Tam giác vuông cân.")
    elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
        print("Loại: Tam giác vuông.")
    else:
        print("Loại: Tam giác thường.")

    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"Chu vi: {a + b + c:.2f}")
    print(f"Diện tích: {s:.2f}")
else:
    print("Không là ba cạnh của một tam giác.")
```

```python
#5
import math

a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

if a == 0:
    # Phương trình bậc nhất
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm.")
        else:
            print("Phương trình vô nghiệm.")
    else:
        x = -c / b
        print(f"Phương trình bậc nhất có nghiệm: x = {x:.2f}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm.")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép: x = {x:.2f}")
    else:
        sqrt_delta = math.sqrt(delta)
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
        print(f"Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {x1:.2f}")
        print(f"x2 = {x2:.2f}")
```

```python
#8
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))

# Tính các định thức
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x:.2f}, y = {y:.2f}")
elif Dx == 0 and Dy == 0:
    print("Hệ phương trình có vô số nghiệm.")
else:
    print("Hệ phương trình vô nghiệm.")
```

```python
#10
x0, y0, r = map(float, input("Nhập x0 y0 r (cách nhau): ").split())
x, y = map(float, input("Nhập tọa độ điểm A x y: ").split())

dist_sq = (x - x0)**2 + (y - y0)**2
r_sq = r**2

if dist_sq == r_sq:
    print("Điểm A nằm trên đường tròn.")
elif dist_sq < r_sq:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```
